## Cálculo do comprimento da caixa central

```python
import numpy as np
```

```python
# Constantes

um = 1e-6
nm = 1e-9
```

```python
# parâmetros do guia 

lambd = 1550*nm
n_c = 1.444 # @1550nm
n_r = 3.476 # @1550nm
k_0 = 2*np.pi/lambd
W_m = 7*um
```

```python
#  effective widths

sigma = 0 # TE mode
# sigma = 1 # TM mode
W_e = W_m + (lambd/np.pi)*((n_c/n_r)**(2*sigma)) * ((n_r**2 - n_c**2)**(-0.5))
W_e
```

```python
neff_app = np.zeros(5)

for i in range(5):
    neff_app[i] = n_r - ((i+1)**2 * np.pi * lambd)/(W_e**2*4*n_r*k_0)
    k_c = (i+1)*np.pi/W_e
    print("neff_app = %1.3f" %(neff_app[i]))
```

```python
L_pi = 2*np.pi/(neff_app[0]-neff_app[2])/k_0
print("O comprimento de batimento: %4.3f um" %(L_pi/um)) 
```
