```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot1 = np.loadtxt("C:/Users/alves/Documents/Photonics/MMI/S_parameters/gain_EME_TE.txt", skiprows=3, delimiter=',')

gain1 = plot1[0:1000,1]
gain2 = plot1[1001:1999,1]
Lambda1 = plot1[0:1000,0]
Lambda2 = plot1[1001:1999,0]
```

```python
# indexes of each port
for i in range(len(plot1)):
    if (plot1[i,0] == 1601.67):
        print(str(i) + '\n' + str(plot1[i,0]))
```

```python
plt.figure(figsize=(8,5))


plt.plot(Lambda1, gain1, lw=1.5, label='Gain $T_3$')
plt.plot(Lambda2, gain2, lw=1.5, label='Gain $T_4$')
plt.xlabel('$\lambda$ (nm)', fontsize=12)
plt.ylabel('Gain (dB)', fontsize=12)
plt.xlim(1520,1575)
plt.title('Gain (dB) - log scale (EME)')
plt.legend(fontsize=12)
plt.grid()
plt.savefig('solver_EME_Gain')
```
